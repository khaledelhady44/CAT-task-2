### filter the data which has (body_style) sedan **and** (engine_type) is ohc

### Delete (price) column and show data without it

```python
automobile.drop("price", axis = 1, inplace = True)
automobile
```

### Show count of missing values in each column

```python
automobile.isna().sum()
```

### Check if the data contain duplicated values or not **and** if it has ,remove these values

```python
automobile.duplicated().sum()
```

```python
automobile = automobile.drop_duplicates()
```

```python
assert automobile.duplicated().sum() == 0
```

### Transform (normalized-losses) from object to numeric

```python
automobile["normalized-losses"] = pd.to_numeric(automobile["normalized-losses"], errors = "coerce")
```

```python
automobile["normalized-losses"].dtype
```

### Replace the missing values in each columns

```python
automobile["normalized-losses"].isna().sum()
```

```python
automobile["normalized-losses"].fillna(0, inplace = True)
```

```python
assert automobile["normalized-losses"].isna().sum() == 0
```

### Save all changes in new excel file called <font color='blue'>cleaned data
